```python
#min(),max() sorted all these are same
s1={1,2,3,3}
s2={1,2}
print(s1.union(s2))
print(s1.intersection(s2))
print(s1.difference(s2))
print(s1.symmetric_difference(s2)) #jo ek dusre me nhi hai
print(s1.isdisjoint(s2))
s1.issubset()
```
