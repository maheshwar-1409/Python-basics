loops

```
import random
jackpot = random.randint(1,100)
x = int(input("guess no"))
c=1
while(x!=jackpot):
  if x<jackpot:
    print("guess higher no")
    c += 1
  else:
    print("guess lower no")
    c += 1
  x = int(input("guess no"))
else:
  print("correct guess")
  print(c,"no. of attempts")
```
